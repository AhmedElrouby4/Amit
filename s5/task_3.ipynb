```python
#                        --------------------------OOP----------------------------

# Task 1 & 2: Fibonacci
class Fibonacci:
    """
    Fibonacci class handles generation of Fibonacci numbers.
    Attributes:
        n (int): Number of terms to generate.
    """
    def __init__(self, n):
        self.n = n

    def fibonacci(self, k):
        """Recursively calculates the k-th Fibonacci number."""
        if k == 0:
            return 0
        elif k == 1:
            return 1
        return self.fibonacci(k-1) + self.fibonacci(k-2)

    def fibonacci_list(self):
        """Generates a list of Fibonacci numbers up to n terms."""
        return [self.fibonacci(i) for i in range(self.n)]

# Example usage
fib_obj = Fibonacci(10)
print("Fibonacci List:", fib_obj.fibonacci_list())


# Task 3 & 4: Set Operations
class SetOperations:
    """
    SetOperations class handles basic set operations like removing elements.
    Attributes:
        input_set (set): The set to perform operations on.
    """
    def __init__(self, input_set):
        self.input_set = input_set

    def remove_item(self, item):
        """Removes a single item from the set."""
        self.input_set.discard(item)  # discard avoids KeyError

    def remove_multiple(self, items):
        """Removes multiple items from the set."""
        self.input_set.difference_update(items)

    def get_set(self):
        return self.input_set

# Example usage
s_obj = SetOperations({1, 2, 3, 4, 5})
s_obj.remove_item(3)
print("After removing 3:", s_obj.get_set())
s_obj = SetOperations({1, 2, 3, 4, 5})
s_obj.remove_multiple({1, 4})
print("After removing 1 and 4:", s_obj.get_set())


# Task 5: Tuple Index
class TupleOperations:
    """
    TupleOperations class handles tuple operations.
    Attributes:
        input_tuple (tuple): The tuple to perform operations on.
    """
    def __init__(self, input_tuple):
        self.input_tuple = input_tuple

    def get_index(self, value):
        """Returns the index of a value in the tuple."""
        return self.input_tuple.index(value)

# Example usage
t_obj = TupleOperations(('apple', 'banana', 'orange', 'grape', 'banana'))
print("Index of 'orange':", t_obj.get_index('orange'))


# Task 6: Split Sentence into Words
class SentenceSplitter:
    """
    SentenceSplitter class splits a sentence into words.
    Attributes:
        sentence (str): The sentence to split.
    """
    def __init__(self, sentence):
        self.sentence = sentence

    def split(self):
        """Splits the sentence into a list of words."""
        return self.sentence.split()

# Example usage
sentence_obj = SentenceSplitter("Hello world, this is Python")
print("Words:", sentence_obj.split())


# Task 7: Palindrome Check
class PalindromeChecker:
    """
    PalindromeChecker class checks if a string is a palindrome.
    Attributes:
        text (str): The string to check.
    """
    def __init__(self, text):
        self.text = text

    def is_palindrome(self):
        """Returns True if text is palindrome, False otherwise."""
        return self.text == self.text[::-1]

# Example usage
pal_obj = PalindromeChecker("radar")
print("Is palindrome?", pal_obj.is_palindrome())


# Task 8: Factorial
class FactorialCalculator:
    """
    FactorialCalculator class calculates factorial of a number.
    Attributes:
        number (int): The number to calculate factorial for.
    """
    def __init__(self, number):
        self.number = number

    def factorial(self):
        """Calculates factorial of the number."""
        result = 1
        for i in range(1, self.number + 1):
            result *= i
        return result

# Example usage
fact_obj = FactorialCalculator(5)
print("Factorial:", fact_obj.factorial())


# Task 9: Squares of Numbers
class SquaresGenerator:
    """
    SquaresGenerator class generates squares of numbers in a range.
    Attributes:
        start (int): Starting number.
        end (int): Ending number.
    """
    def __init__(self, start=1, end=10):
        self.start = start
        self.end = end

    def generate_squares(self):
        """Generates a list of squares from start to end."""
        return [x**2 for x in range(self.start, self.end + 1)]

# Example usage
sq_obj = SquaresGenerator()
print("Squares:", sq_obj.generate_squares())
```
